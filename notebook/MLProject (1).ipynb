```python
from sklearn.preprocessing import MinMaxScaler

# Identify numerical columns (excluding 'date' if it's a datetime object)
# Assuming 'date' is already converted to datetime as in your previous code
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
if 'date' in numerical_cols:
    numerical_cols.remove('date')

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply normalization to the numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Display the first few rows to see the normalized data
print("DataFrame after MinMaxScaler Normalization:")
print(df.head())
```

```python
#Feature engineering
df['price_lag_1'] = df['price'].shift(1)
df['price_lag_7'] = df['price'].shift(2) # Example for a weekly lag
```

```python
#rolling window
df['price_rolling_mean_7'] = df['price'].rolling(window=7).mean()
df['price_rolling_std_7'] = df['price'].rolling(window=7).std()
```

```python
#tume based features
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['dayofweek'] = df['date'].dt.dayofweek
df['dayofyear'] = df['date'].dt.dayofyear
# You can also add features like 'weekofyear', 'quarter', etc.
```

```python
#Interaction features
df['price_diff_rolling_mean_7'] = df['price'] - df['price_rolling_mean_7']
```

```python
# Assuming 'df' is your combined and feature-engineered DataFrame

# 1. Define features (X) and target (y)
features = ['24h_volume', 'price_lag_1', 'price_rolling_mean_7', 'year', 'month', 'dayofweek'] # Add your engineered features
target = 'price'

# Split data into features (X) and target (y)
X = df[features]
y = df[target]


# 2. Split data chronologically
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state = 1, shuffle=False)

# 3. Scale Features
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Fit scaler on training data ONLY
X_train_scaled = scaler.fit_transform(X_train)

# Transform both training and testing data
X_test_scaled = scaler.transform(X_test)

# Convert scaled arrays back to DataFrames (optional, but keeps column names)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=features, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=features, index=X_test.index)


# 4. Model Training (Example using RandomForestRegressor)
from sklearn.ensemble import RandomForestRegressor
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import GridSearchCV,TimeSeriesSplit
from sklearn.pipeline import Pipeline # Import Pipeline
import numpy as np

# --- 6. Create a Pipeline ---
# The pipeline includes scaling and the model
pipeline = Pipeline([
    ('scaler', MinMaxScaler()),         # Step 1: Scale the features
    ('regressor', RandomForestRegressor(random_state=42)), # Step 2: The RandomForestRegressor model
    ])

# --- 7. Hyperparameter Tuning (using TimeSeriesSplit and GridSearchCV) ---
param_grid = {
    'regressor__n_estimators': [50, 100, 200],
    'regressor__max_depth': [None, 10, 20],
    'regressor__min_samples_split': [2, 5],
    'regressor__min_samples_leaf': [1, 2]
}

# Use TimeSeriesSplit for cross-validation
tscv = TimeSeriesSplit(n_splits=5) # Adjust the number of splits as needed

grid_search = GridSearchCV(estimator=pipeline, # Pass the pipeline to GridSearchCV
                           param_grid=param_grid,
                           cv=tscv,
                           scoring='neg_mean_squared_error', # Optimize for MSE (negated)
                           n_jobs=-1)

# Fit the grid search on the training data
grid_search.fit(X_train, y_train) # Pass the original X_train (pipeline handles scaling)

# Get the best pipeline (with best hyperparameters)
best_pipeline = grid_search.best_estimator_

print("Best Hyperparameters:", grid_search.best_params_)

# --- 8. Prediction ---
# Use the best pipeline to make predictions on the test data
predictions = best_pipeline.predict(X_test)

# --- 9. Evaluation ---
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, predictions)

print(f"\nMSE with best pipeline: {mse}")
print(f"RMSE with best pipeline: {rmse}")
print(f"MAE with best pipeline: {mae}")

# --- 10. Feature Importance (from the best model within the pipeline) ---
# Access the regressor from the best pipeline
if hasattr(best_pipeline.named_steps['regressor'], 'feature_importances_'):
    importances = best_pipeline.named_steps['regressor'].feature_importances_
    feature_importance_df = pd.DataFrame({'feature': features, 'importance': importances})
    feature_importance_df = feature_importance_df.sort_values('importance', ascending=False)
    print("\nFeature Importances (from best model):")
    print(feature_importance_df)
```
