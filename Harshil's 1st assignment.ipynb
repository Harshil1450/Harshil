3. Compare and contrast mutable and immutable objects in Python with
examples.
Ans.
In Python, mutable objects can be changed after creation, while
immutable objects cannot be modified once created.

Mutable:

Examples: Lists, dictionaries, sets.

Behavior: Elements can be added, removed, or modified.

```python
my_list = [1, 2, 3]
my_list[0] = 5
print(my_list)
```

Immutable:
Examples: Strings, tuples, integers.

Behavior: Cannot change the object itself, but can reassign the variable to
a new object.

```python
my_string = "hello"
new_string = "H" + my_string[1:]
print(new_string)
```

Key difference: Mutables allow in-place modifications, while immutables
require creating a new object for changes.

5. Explain the concept of type casting in Python with examples.
Ans.
Type casting in Python refers to converting one data type into another. It can
be done using built-in functions.
Types of Type Casting:
Implicit Casting: Automatically done by Python when compatible types are
involved.

```python
x = 5
y = 2.0
result = x + y
print(result)
```

Explicit Casting: Manually converting a type using functions like int(),
float(), str(), etc.

```python
x = "10"
y = int(x)
print(y + 5)
```

Explicit casting is necessary when automatic conversion isn’t possible or
could lead to errors.

6. How do conditional statements work in Python? Illustrate with examples.

Ans.
In Python, conditional statements are used to execute code based on specific
conditions. The main conditional statements are if, elif, and else.
Syntax:
if: Executes the block if the condition is true.
elif: (else if) Tests another condition if the previous if is false.
else: Executes if none of the above conditions are true

```python
x = 10
if x > 15:
 print("Greater than 15")
elif x == 10:
 print("Equal to 10")
else:
 print("Less than 10")
```

Conditional statements help control the flow of the program based on logic.

Ans.
Python has two main types of loops: for loops and while loops.
1. for Loop
Used for iterating over a sequence (like lists, strings, or ranges).
Use case: When you know the number of iterations in advance.

```python
for i in range(5):
 print(i)
```

while Loop
Executes as long as a condition is true.
Use case: When the number of iterations is uncertain and depends on a
condition.


```python
x = 0
while x < 5:
 print(x)
 x =x + 1
```

Control Statements:
break: Exit the loop prematurely.
continue: Skip the current iteration and continue with the next one.
These loops are used for repetitive tasks, iterating over data, and performing
operations until a condition is met.
